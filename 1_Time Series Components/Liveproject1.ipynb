## Detect components

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("stock_prices.csv")
```

```python
df["date"] = pd.to_datetime(df["timestamp"])
df = df.set_index('date').asfreq('D')
df = df.dropna()
```

```python
def plot_components(timeseries):
    result = seasonal_decompose(timeseries, model='ad', period=int(len(df)/2))
    plt.rcParams.update({'figure.figsize': (10,10)})
    result.plot()
    plt.show()
```

```python
plot_components(df["T"])
```

```python
plot_components(df["VZ"])
```
